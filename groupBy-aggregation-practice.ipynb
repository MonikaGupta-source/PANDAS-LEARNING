```python
import pandas as pd
```

```python
data = {
    "Department": ["HR", "IT", "Finance", "IT", "HR", "Finance", "IT", "HR", "Finance", "IT"],
    "Employee": ["Monika", "Rahul", "Sneha", "Daksh", "Ragini", "Amit", "Priya", "Vikas", "Neha", "Rohit"],
    "Salary": [45000, 72000, 60000, 85000, 48000, 55000, 90000, 42000, 67000, 78000],
    "Age": [25, 28, 32, 24, 30, 35, 27, 40, 29, 26],
    "City": ["Delhi", "Mumbai", "Bangalore", "Delhi", "Hyderabad", "Mumbai", "Pune", "Delhi", "Bangalore", "Mumbai"]
}

df = pd.DataFrame(data)
```

```python
df
```

```python
print(df.groupby('Department')['Salary'].mean())
print(df.groupby('Department')['Age'].max())
```

```python
df.groupby('Department').agg({
    "Salary" : ['mean','min','max'],
    "Age"  : ['max']
})
```

```python
pivot = df.pivot_table(values=['Salary'],index='City',columns='Department',aggfunc='mean')
pivot
```

```python
pivot['Salary'] =pivot['Salary'].fillna(pivot['Salary'].mean())
pivot
```
