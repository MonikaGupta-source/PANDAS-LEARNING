```python
import pandas as pd
import numpy as np

# 1. Create a messy DataFrame
data = {
    "Name": ["Alice", "Bob", "Chrlie", "David", "Alice", None],
    "Age": ["25", "30", None, "40", "25", "30"],      # ages stored as strings, with missing value
    "City": ["Delhi", "Mumbai", "Pun", "Bangalore", "Delhi", "Mumbai"],  # typo: "Pun" -> "Pune"
    "Score": [85, None, 78, 88, 85, 91],              # missing value in Score
    "Extra": ["x", "y", "z", "x", "y", "z"]           # extra column we don't really need
}

df = pd.DataFrame(data)

print("Original messy DataFrame:")
print(df)
```

**checkout the info of your dataframe**

```python
print(f"The total column of your data : {len(df.columns)}")
print(f"The total rows of your data : {len(df)}")
print(f"The total info of the dataframe is : {df.info}")


print(f"Total null values in the data is : {df.isna().sum()}")
```

**copy the orginal data then make changes in the copied**

```python
copied = df.copy()
copied
```

**fill the null values**

```python
copied['Name'] = copied['Name'].fillna("Unknown")
copied
```

```python
copied['Age'].dtypes
```

```python
copied['Age'] = copied['Age'].fillna(0)
print(copied)
copied['Age'].dtype
```

```python
copied['Age'] = copied['Age'].astype(int)
copied['Age'].dtype
```

```python
copied['Age'] = copied['Age'].replace({0:copied['Age'].mean()})
copied
```

```python
copied['Score'].dtype
```

```python
copied['Score'] = copied['Score'].fillna(copied['Score'].mean())
copied
```

```python
copied
```

```python
copied.drop('Extra',axis=1,inplace=True)
copied
```

```python
copied['City'] = copied['City'].replace({"Pun":"Pune"})
copied
```
